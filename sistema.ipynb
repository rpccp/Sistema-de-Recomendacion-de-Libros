```python
import pandas as pd

# Cargo el dataframe
df = pd.read_csv('books.csv')
df
```

```python
# De cada fila, elijo las columnas que voy a tener en cuenta para el sistema, tomo el contenido y lo convierto en un string

def representacion_textual(fila):
    representacion_textual = f"""Titulo: {fila['title']}
Autores: {fila['authors']}
Descripcion: {fila['description']}
Categorias: {fila['categories']}
Anio de Publicacion: {fila['published_year']}
Rating Promedio: {fila['average_rating']}
Numero de Paginas: {fila['num_pages']}
"""
    return representacion_textual
```

```python
# Aplico la funcion a cada fila

df['representacion_textual'] = df.apply(representacion_textual, axis=1)
```

```python
import faiss
import requests
import numpy as np

dim = 3584

indice = faiss.IndexFlatL2(dim)

# Inicializamos input lleno de ceros

X = np.zeros((len(df['representacion_textual']), dim), dtype='float32')
```

```python


for i, representacion in enumerate(df['representacion_textual']):

    # Toma la representacion y hace una request, uso el localhost de Ollama

    respuesta = requests.post('http://localhost:11434/api/embeddings',
                              json = {
                                  'model': 'deepseek-r1:7b', 
                                  'prompt': representacion
                              })
    embedding = respuesta.json()['embedding']

    X[i] = np.array(embedding)

indice.add(X)
```
