```python
import pandas as pd
df = pd.read_csv('books.csv')
df
```

```python
def representacion_textual(columna):
    representacion_textual = f"""Titulo: {columna['title']}
Autores: {columna['authors']}
Descripcion: {columna['description']}
Categorias: {columna['categories']}
Anio de Publicacion: {columna['published_year']}
Rating Promedio: {columna['average_rating']}
Numero de Paginas: {columna['num_pages']}
"""
    return representacion_textual
```

```python
df['representacion_textual'] = df.apply(representacion_textual, axis=1)
```

```python
import faiss
import requests
import numpy as np

dim = 4096

indice = faiss.IndexFlatL2(dim)

X = np.zeros((len(df['representacion_textual']), dim), dtype='float32')
```

```python
for i, representacion in enumerate(df['representacion_textual']):
    if i % 100 == 0:
        print(i)


    respuesta = requests.post('http//localhost:11434/api/embeddings',
                              json = {
                                  'model': 'llama2', 
                                  'prompt': representacion
                              })
    embedding = respuesta.json()['embedding']

    X[i] = np.array(embedding)

indice.add(X)
```
